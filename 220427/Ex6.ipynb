```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('data7_data3.npy')
data_2d=data.reshape(-1,100*100)
data_2d.shape
```

```python
from sklearn.cluster import KMeans
X = data_2d
km = KMeans(n_clusters=3)
km.fit(X)
```

```python
km.labels_
```

```python
np.unique(km.labels_)
```

```python
import matplotlib.pyplot as plt
def f(data,r=1):
    n=len(data)
    rows=int(np.ceil(n/10))
    cols = n if rows < 2 else 10
    fig,axs = plt.subplots(rows,cols,figsize = (rows*r,cols*r))
    for i in range(rows):
        for j in range(cols):
            if i*10+j<n:
                axs[i,j].imshow(data[i*10+j],cmap = 'gray_r')
            axs[i,j].axis('off')
    plt.show()
```

```python
f(data[km.labels_ == 0])
```

```python
f(data[km.labels_ == 1])
```

```python
f(data[km.labels_ == 2])
```
