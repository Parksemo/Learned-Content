```python
from konlpy.corpus import kolaw
t_data = kolaw.open('constitution.txt')
t_data.name
```

```python
from soynlp import DoublespaceLineCorpus
t_data = DoublespaceLineCorpus(t_data.name,iter_sent=True)
```

```python
from konlpy.tag import Okt
from tensorflow.keras.preprocessing.text import Tokenizer
tw1 = Okt()
```

```python
s_data =[]
for w in t_data:
    d=tw1.nouns(w)
    d=[i for i in d if len(i)>1]
    s_data.append(d)
s_data
```

```python
ck_t=Tokenizer()
ck_t.fit_on_texts(s_data)
ck_t.index_word
```

```python
encoded = ck_t.texts_to_sequences(s_data)
encoded
```

```python
ck_t.word_counts
```

```python
max_l = max(len(x)for x in encoded)
max_l
```

```python
import numpy as np
for i in encoded:
    while len(i)<max_l:
        i.append(0)
data=np.array(encoded)
data
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
end_data=pad_sequences(encoded)
end_data
```

```python
end_data = pad_sequences(encoded,padding='post',maxlen=5)
end_data
```

```python
end_data = pad_sequences(encoded,padding='post',truncating='post',maxlen=5)
end_data
```
