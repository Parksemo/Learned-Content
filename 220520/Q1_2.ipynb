```python
import pickle
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,Embedding, Dense, LSTM,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard
import re
from konlpy.tag import Okt
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
with open("data.pickle","rb") as fr:
    전처리_된_데이터 = pickle.load(fr)
```

```python
전처리_된_데이터.keys()
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4,restore_best_weights=True)
t_bd=TensorBoard(log_dir="logs")
```

```python
mc1 = ModelCheckpoint('best_m1.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
model1 = Sequential()
model1.add(Embedding(전처리_된_데이터['단어수'], 256))
model1.add(LSTM(128,dropout=0.5,return_sequences=True))
model1.add(BatchNormalization())
model1.add(LSTM(64,dropout=0.5))
model1.add(Dense(1, activation='sigmoid'))
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history1 = model1.fit(전처리_된_데이터['학습_데이터'],전처리_된_데이터['학습_결과'], epochs=15, callbacks=[es, mc1,t_bd], batch_size=128,validation_data=(전처리_된_데이터['검증_데이터'],전처리_된_데이터['검증_결과']))
```

```python
epochs1 = range(1, len(history1.history['acc']) + 1)
plt.plot(epochs1, history1.history['loss'])
plt.plot(epochs1, history1.history['val_loss'])
plt.title('m1 loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
loaded_model1 = load_model('best_m1.h5')
a1 = loaded_model1.evaluate(전처리_된_데이터['테스트_데이터'],전처리_된_데이터['테스트_결과'])[1]
a1
```

```python
def 문장_전처리(문장):
    okt = Okt()
    문장 = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', 문장)
    문장 = okt.morphs(문장, stem=True)
    문장 = [w for w in 문장 if not w in 전처리_된_데이터['불용어']] 
    encoded = 전처리_된_데이터['토큰'].texts_to_sequences([문장])
    pad_new = pad_sequences(encoded, maxlen = 전처리_된_데이터['입력길이'])
    score = float(loaded_model1.predict(pad_new))
    if(score > 0.5):
        print(f"{score * 100:.2f}% 확률로 긍정 게임 리뷰입니다.\n")
    else:
        print(f"{(1 - score) * 100:.2f}% 확률로 부정 게임 리뷰입니다.\n")
```
