```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
import pandas as pd
from pandas import DataFrame
df=pd.read_excel('data1.xlsx')
```

```python
df_1=pd.read_excel('data1.xlsx',header=4)
df_1=df_1.drop(0)
df_1.rename(columns={'매출액':'매출액점유율'}, inplace=True)
df_1=df_1.reset_index(drop=True)
for i in range(len(df_1)):
    if str(df_1.loc[i][2]) >= '2020-01-01 00:00:00':
        df_1=df_1.drop(i)
    elif str(df_1.loc[i][2]) < '2019-01-01 00:00:00':
        df_1=df_1.drop(i)
df_1=df_1.reset_index(drop=True)
df_1
```

2. 개봉일 기준 월별 매출순위(1~10)를 막대그래프와 꺽은선 그래프로 같이 표시하시오

```python
df_2=df_1
df_2_1=df_2
a=[]
b=[]    

def 추출(c):

    for i in range(len(c)):
        if str(c.loc[i][2]) >= '2019-02-01 00:00:00':
            c=c.drop(i)
        elif str(c.loc[i][2]) < '2019-01-01 00:00:00':
            c=c.drop(i)
    c=c.reset_index(drop=True)
    c


    for i in range(len(c)):
        a.append(c.loc[i][1])
        b.append(c.loc[i][3])

    data={
        '영화명':a,
        '매출액':b
    }
    df_t=DataFrame(data)
    df_t
    df_t=df_t.sort_values(by='매출액',ascending=False)
    df_t
    for i in range(10,len(df_t)):
        df_t=df_t.drop(i)
    df_t
    df_t=df_t.reset_index(drop=True)
    df_t
    plt.plot(df_t['영화명'],df_t['매출액'],'r*-')
    plt.bar(df_t['영화명'],df_t['매출액'])
    plt.xticks(rotation=90)
    plt.show()

    
추출(df_2_1)
```
