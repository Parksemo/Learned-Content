## Q 박스오피스 data 분석

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

1. 박스오피스 2019년 한국 박스오피스 data를 엑셀로 받고 데이터프레임을 완성하시오

```python
df=pd.read_excel('data1.xlsx',skiprows=4)
df.columns=['순위','영화명','개봉일','매출액','점유율','누적매출액','관객수','누적관객수','스크린수','상영횟수','대표국적','국적','제작사','배급사','등급','장르','감독','배우']
df=df.loc[:,['영화명','개봉일','매출액','점유율','누적매출액','관객수','누적관객수','스크린수','상영횟수']]
df=df.astype({'개봉일':str})
df=df[df.개봉일.str.contains('2019-',case=True)]
df
```

2. 개봉일 기준 월별 매출순위(1~10)를 막대그래프와 꺽은선 그래프로 같이 표시하시오

```python
for i in range(1,13):
    if i>=10:
        df2=df[df.개봉일.str.contains(f'-{i}-',case=True)]
    else:
        df2=df[df.개봉일.str.contains(f'-0{i}-',case=True)]
    df2=df2.sort_values(by='매출액',ascending=False).head(10)
    plt.bar(df2.영화명, df2.매출액,color='b')
    plt.plot(df2.영화명, df2.매출액,'r--')
    plt.title(str(i)+'월')
    plt.xticks(rotation=90)
    plt.show()
```

3. 스크린수 대비 매출액을 산출하여 1~10위까지를 시각화 하시오

```python
df3=df.copy()
df3['스크린수대비매출액']=(df3['매출액']/df3['스크린수'])
df3=df3.sort_values(by='스크린수대비매출액', ascending=False).head(10)
plt.bar(df3.영화명, df3.스크린수대비매출액)
plt.plot(df3.영화명, df3.스크린수대비매출액,'r--')
plt.xticks(rotation=90)
plt.show()
```

4. 상영횟수 대비 관객수를 계산하여 1회 상영시 평균 관객수를 이용하여 1~10위 를 시각화 하시오

```python
df4=df.copy()
df4['상영횟수_대비_관객수']=(df4['관객수']//df4['상영횟수'])
df4=df4.sort_values(by='상영횟수_대비_관객수', ascending=False).head(10)
plt.bar(df4.영화명, df4.상영횟수_대비_관객수)
plt.plot(df4.영화명, df4.상영횟수_대비_관객수,'r--')
plt.xticks(rotation=90)
plt.show()
```

5. 관객수 대비 누적매출액을 계산하여 평균 1명당 영화 관람에 사용한 금액을 계산하여 1~10위까지의 순위를 시각화 하시오

```python
df5=df.copy()
df5['관객수_대비_누적매출액']=(df5['누적매출액']/df5['관객수'])
df5=df5.sort_values(by='관객수_대비_누적매출액', ascending=False).head(10)
plt.bar(df5.영화명, df5.관객수_대비_누적매출액)
plt.plot(df5.영화명, df5.관객수_대비_누적매출액,'r--')
plt.xticks(rotation=90)
plt.show()
```

6. 스크린수 대비 상영횟수를 계산하여 평균 1개의 스크린에서 상영한 횟수를 통한 1~10위 순위를 시각화 하시오

```python
df6=df.copy()
df6['스크린수_대비_상영횟수']=(df6['상영횟수']/df6['스크린수'])
df6=df6.sort_values(by='스크린수_대비_상영횟수', ascending=False).head(10)
plt.bar(df6.영화명, df6.스크린수_대비_상영횟수)
plt.plot(df6.영화명, df6.스크린수_대비_상영횟수,'r--')
plt.xticks(rotation=90)
plt.show()
```

7. 매출액 점유 기준 1~10위 를 선별하고 원그래프를 만드시오

```python
df7 = df.sort_values(by='매출액', ascending=False).head(10)
plt.pie(df7.매출액, labels=df7.영화명, autopct='%.2f%%')
plt.show()
```

8. 위 문제에서 만든 data에서 5위까지의 점유를 제외한 6위~10위 까지의 점유율을 원그래프로 나타내시오

```python
df8 = df.sort_values(by='매출액', ascending=False).head(10)
plt.pie(df8.loc[5:,'매출액'], labels=df8.loc[5:,'영화명'], autopct='%.2f%%')
plt.show()
```
