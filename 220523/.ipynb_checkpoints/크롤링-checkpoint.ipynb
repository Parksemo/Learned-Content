```python
import time
import csv
import requests
from bs4 import BeautifulSoup
url=("https://movie.naver.com/movie/point/af/list.naver?&page=")
in_data=[]

for page in range (1,1001):
    print(f"page{page} 크롤링중")
    r=requests.get(url+str(page))
    soup=BeautifulSoup(r.text,"html.parser")
    data=soup.find_all("td",attrs={"class":"title"})
    for i in data:
        if i.a:
            in_data.append([i.a.text, i.em.text, i.br.next_sibling.strip()])
    time.sleep(0.1)

with open("movie_data.csv","w",encoding="utf-8-sig",newline='') as f:
    w=csv.writer(f)
    w.writerow(["영화명","평점","리뷰"])
    w.writerows(in_data)

with open("movie_data.csv","r",encoding="utf-8-sig",newline='') as f:
    r=csv.reader(f)
    n=True
    for i in r:
        if n:
            n=False
            continue
```
