```python
import pandas as pd
import pickle
```

```python
with open('df_sector.pickle','rb') as f:
    df_sector = pickle.load(f)
```

```python
with open('sector.pickle','rb') as f:
    sector = pickle.load(f)
```

```python
#등락률
for sector_name, df_dict in df_sector.items():
    for stockcode, df in df_dict.items():
        등락률 = []
        for i in range(len(df)):
            if i == len(df)-1:
                break
            else:    
                등락률.append((df['종가'][i]-df['종가'][i+1])/df['종가'][i+1]*100)
        등락률.append(0)
        df['등락률']=등락률
```

```python
#거래변동량
for sector_name, df_dict in df_sector.items():
    for stockcode, df in df_dict.items():
        거래변동량 = []
        for i in range(len(df['거래량'])):
            if i == len(df['거래량'])-1:
                break
            else:    
                거래변동량.append(abs((df['거래량'][i]-df['거래량'][i+1])))
        거래변동량.append(0)
        df['거래변동량']=거래변동량
```

```python
#종가정규화
from sklearn.preprocessing import MinMaxScaler
for sector_name, df_dict in df_sector.items():
    for stockcode, df in df_dict.items():
        scaler = MinMaxScaler()
        scaler.fit(pd.DataFrame(data = df['종가']))
        df_scaled = scaler.transform(pd.DataFrame(data = df['종가']))
        df['정규화_종가']=pd.DataFrame(data = df_scaled)
```

```python
for sector_name, df_dict in df_sector.items():
    for stockcode, df in df_dict.items():
        df = df[['일자','등락률','거래변동량','정규화_종가']]
        df_sector[sector_name][stockcode] = df
```

```python
df_pr_sector = df_sector
```
