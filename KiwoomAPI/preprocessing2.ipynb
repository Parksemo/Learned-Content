```python
d = {}
for sector_name, df_dict in df_pr_sector.items():
    dd  = pd.DataFrame(data = df_pr_sector['Food']['097950']['일자'])
    ddd_등락률 = pd.DataFrame(data = df_pr_sector['Food']['097950']['일자'])
    ddd_거래변동량 = pd.DataFrame(data = df_pr_sector['Food']['097950']['일자'])
    ddd_정규화_종가 = pd.DataFrame(data = df_pr_sector['Food']['097950']['일자'])
    for stockcode, df in df_dict.items():
        ddd_등락률[f'{stockcode}등락률']=df['등락률']
        ddd_거래변동량[f'{stockcode}거래변동량']=df['거래변동량']
        ddd_정규화_종가[f'{stockcode}정규화_종가']=df['정규화_종가']
    del(ddd_등락률['일자'])
    del(ddd_거래변동량['일자'])
    del(ddd_정규화_종가['일자'])
    print(ddd_등락률)
    print(ddd_거래변동량)
    print(ddd_정규화_종가)
    dd['등락률']=ddd_등락률.mean(axis='columns')
    dd['거래변동량']=ddd_거래변동량.mean(axis='columns')
    dd['정규화_종가']=ddd_정규화_종가.mean(axis='columns')
    print(dd)
    d[f'{sector_name}']=dd
```
