데이터 수집

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def data_gen(n,b,v):
    x = np.zeros(shape=(n,2))
    y = np.zeros(shape=n)
    
    for i in range(0,n):
        x[i][0]=1
        x[i][1]=i
        y[i] = (i+b) + random.uniform(0,1)*v
    return x,y

x,y=data_gen(100,25,10)
plt.plot(x[:,1],y,'o')
```

```python
def g_descent(x,y,th,ap,m,ns): #th:세타(정답위치?) m:입력된 data의 개수 ns:반복수
    xT=x.T #np.transpose(x), x.transpose(),x.T
    th_l=[]
    cost_l=[]
    for i in range(0,ns):
        h = np.dot(x,th) #np.dot(x,th), x.dot(th)
        loss = h-y #오차생성
        cost = np.sum(loss**2)/(2*m)
        g=np.dot(xT,loss)/m
        th=th - ap*g#th세타 ap 학습률 g 기울기
        if i %250 == 0:
            th_l.append(th)
        cost_l.append(cost)
    return th,np.array(th_l),cost_l
m,n =np.shape(x)
ns= 5000#반복수
ap= 0.0005
th= np.ones(n)
th,th_l,cost_l=g_descent(x,y,th,ap,m,ns)
y_p_step = np.dot(x,th_l.T)
plt.plot(x[:,1],y,'o')
for i in range(0,20,5):
    plt.plot(x[:,1],y_p_step[:,i],label=f'L{i}')
plt.legend()
plt.show()
```

```python
y_p = np.dot(x,th)
plt.plot(x[:,1],y,'o')
plt.plot(x[:,1],y_p)
```
