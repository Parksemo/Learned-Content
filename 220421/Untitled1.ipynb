경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10,10,1)
fx=x**2#y=x**2

plt.plot(x,fx)
```

```python
x = np.arange(-10,11,1)
fx=x**2#y=x**2

x_new = 10
d=[] #변화된 x기록
y=[] #y기록
learning_rate=0.1 #a learning_rate 학습률
for i in range(100):
    x_old=x_new
    x_new=x_old-learning_rate*2*x_old
    d.append(x_new)
    y.append(x_new**2)
plt.plot(x,fx) #곡선
plt.scatter(d,y)
plt.plot(d,y)
```
