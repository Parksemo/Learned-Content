결정 트리

```python
import pandas as pd
wine = pd.read_csv('day5_data4.csv')
wine
```

```python
wine.describe()
```

```python
X = wine[['alcohol','sugar','pH']].to_numpy()
Y = wine['class'].to_numpy()
from sklearn.preprocessing import StandardScaler
x_data = StandardScaler().fit_transform(X)
from sklearn.model_selection import train_test_split
t_x,tt_x,t_y,tt_y = train_test_split(x_data, Y, random_state=42)
```

```python
t_x.shape,t_y.shape
```

```python
from sklearn.linear_model import LogisticRegression
lo_g = LogisticRegression()
lo_g.fit(t_x,t_y)
lo_g.score(t_x,t_y),lo_g.score(tt_x,tt_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(t_x,t_y)
dt.score(t_x,t_y),dt.score(tt_x,tt_y)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plot_tree(dt,max_depth=5,filled=True,feature_names=['alcohol','sugar','pH'])
plt.show()
```

```python
dt1 = DecisionTreeClassifier(max_depth=5,random_state=42) #max_depth 5층까지만
dt1.fit(t_x,t_y)
dt1.score(t_x,t_y),dt1.score(tt_x,tt_y)
```

```python
plt.figure(figsize=(10,7))
plot_tree(dt1,max_depth=5,filled=True,feature_names=['alcohol','sugar','pH'])
plt.show()
```
