1. data 수집

```python
import pandas as pd
df = pd.read_csv('day5_data1.csv')
df
```

2. data 전처리 1. 필요유무에 따른 정리

```python
# 필용없는 정보 제거
df.pop('who')
df.pop('Country')
df.pop('Years on Internet')
df.dtypes
```

```python
#범주화 (원핫인코딩사용을 위해)
ck_c=['Gender','Household Income','Sexual Preference',
      'Education Attainment','Major Occupation','Marital Status']
for i in ck_c:
    df[i] = df[i].astype('category') #category : 범주형data
df.dtypes
```

```python
df_one_hot=pd.get_dummies(df) #범주화된 data만 원핫인코딩됨
df_one_hot
```

2. data 전처리 2. 결측치

```python
df_one_hot.isnull().sum() #결측치, Age 결측치 존재
```

```python
df_one_hot.loc[pd.isnull(df_one_hot['Age']),'Age']=df_one_hot['Age'].mean()
```

3. 입력 data정리

```python
X=df_one_hot.iloc[:,1:].values
Y=df_one_hot.iloc[:,0].values.reshape(-1,1)
X.shape,Y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler #전처리
m_m_s=MinMaxScaler()
X_data=m_m_s.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
t_x,tt_x,t_y,tt_y=train_test_split(X_data,Y,test_size=0.3,random_state=42)
```

4. 모델 생성 및 학습

```python
from sklearn.linear_model import LogisticRegression
lo_g=LogisticRegression(fit_intercept=True)
lo_g.fit(t_x,t_y.flatten())
```

5. 테스트 및 검증

```python
lo_g.predict(tt_x[:5])
```

```python
lo_g.predict_proba(tt_x[:5]) #predict_proba 0과 1의 확률 0.5업다운
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
y_t = tt_y.copy()
y_p = lo_g.predict(tt_x)
confusion_matrix(y_t,y_p)
```

```python
accuracy_score(y_t,y_p)
```
