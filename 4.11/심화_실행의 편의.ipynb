그래프영역표시

```python
x=[1,2,3,4]
y=[4,5,6,7]
#plt.fill_between(x의 좌표 범위, y의 좌표 범위, 명도)
#plt.fill#영역표시
#plt.fill_between#수평영역
#plt.fill_betweenx#수직영역
plt.plot(x,y)
#plt.fill_between(x[1:3],y[1:3],alpha=0.5)
plt.fill_betweenx(y[1:3],x[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]
plt.plot(x,y)
plt.plot(x,y1)
plt.fill_between(x[1:3],y[1:3],y1[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]
plt.plot(x,y)
plt.plot(x,y1)
plt.fill([2.0,2.0,3.5,3.5],[2.0,6.0,6.0,3.0],alpha=0.5)
```

축크기

```python
import numpy as np
x=np.linspace(-10,10,100)
y=x**3
plt.plot(x,y)
plt.xscale('symlog')
#plt.ysclae('log')
```

```python
x=np.linspace(0,10,100)
y=x**3
plt.plot(x,y)
plt.yscale('log')
```

```python
x=np.linspace(0,10,100)
y=np.exp(x)
plt.plot(x,y)
plt.yscale('log')
```
