```python
review_l = []
while True:
# for k in range(2):
    try:
        for j in range(1,11):
            driver.find_element_by_xpath(f'//*[@id="premium-pagenation-wrap"]/div[1]/ul/li[{j}]/a').click()
            time.sleep(1)
            review_t = driver.find_elements_by_css_selector('.uxelayer_ctrl .con')
            for i in range(0, len(review_t)):
                review_l.append(review_t[i].text)
        driver.find_element_by_xpath(f'//*[@id="premium-pagenation-wrap"]/div[1]/a[3]').click()
        time.sleep(1)
    except:
        break
review_l
```

```python
len(review_l)
```

```python
df = pd.DataFrame(review_l, columns = ['리뷰'])
df
```

```python
df.isnull().values.any()
```

```python
df['리뷰'].nunique()
```

```python
import numpy as np
```

```python
df=df.drop_duplicates(subset=['리뷰'])
df['리뷰']=df.리뷰.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','') 
df['리뷰']=df.리뷰.str.replace('^ +','')
df['리뷰']=df.리뷰.replace('',np.nan)
df=df.dropna(how='any')
```

```python
print('전처리 후 테스트용 샘플의 개수 :',len(df))
```

```python
df.to_csv("선풍기_리뷰.csv", encoding='utf-8-sig')
```
