```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
from sklearn.model_selection import train_test_split
import numpy as np
from konlpy.tag import Okt
# from tensorflow.keras.preprocessing.text import Tokenizer 
```

```python
data = pd.read_table('ratings_total.txt', names=['ratings','reviews'])
```

```python
data
```

```python
data.ratings
```

```python
data=data.drop_duplicates(subset=['reviews'])
data['reviews']=data.reviews.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','')
data['reviews']=data.reviews.str.replace('^ +','')
data['reviews']=data.reviews.replace('',np.nan)
data=data.dropna(how='any')
s_w=['은','는','이','가','를','들','에게','의','을','도','으로','만','라서','하다']
```

```python
okt=Okt()
from tqdm import tqdm 
X_data=[]
for i in tqdm(data['reviews']):
    tk_d=okt.morphs(i)
    end_d=[w for w in tk_d if not w in s_w]
    X_data.append(end_d)
X_data
```
