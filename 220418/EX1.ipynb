## 시몬(seaborn)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.set_style('whitegrid')
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
fmri.sample(n=10,random_state=1)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri) #hue 항목 분류
```

```python
t=sns.load_dataset('tips')
sns.regplot(x='total_bill',y='tip',data=t) #regplot회귀그래프
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data=t) #scatterplot산점도
```

```python
sns.countplot(x='smoker',hue='time',data=t) #countplot 수치비교그래프
```

```python
t.sample(n=10,random_state=1)
```

```python
sns.barplot(x='day',y='total_bill',data=t) #barplot 막대그래프
```

```python
# 분포를 나타내는 그래프 - 바이올릿플롯, 스웜플롯
sns.violinplot(x='day',y='total_bill',hue='smoker',data=t) #violinplot
```

```python
sns.swarmplot(x='day',y='total_bill',hue='smoker',data=t)
```

```python
g=sns.FacetGrid(t,col='time',row='sex') #패싯그리드FacetGrid 그래프의 틀만제공
```

```python
g=sns.FacetGrid(t,col='time',row='sex')
g.map(sns.scatterplot,'total_bill','tip') #map
```

```python
g=sns.FacetGrid(t,col='time',row='sex')
g.map_dataframe(sns.histplot,x='total_bill') #histplot 막대그래프
```
