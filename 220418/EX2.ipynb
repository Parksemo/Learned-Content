```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#이전에 배운 한글표시와 음수표시와 동일한 코드
#matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rc('font',family='Malgun Gothic')
#matplotlib.rcParams['axes.uncode_minus']=False
matplotlib.rc('axes',unicode_minus=False)
```

```python
붓꽃data=sns.load_dataset('iris')
타이타닉data=sns.load_dataset('titanic')
팁data=sns.load_dataset('tips')
운송data=sns.load_dataset('flights')
```

```python
붓꽃data
```

```python
x=붓꽃data.petal_length.values
sns.rugplot(x)
plt.title("붓꽃data중 꽃잎 길이 러그 플롯") 
plt.show()
```

```python
sns.kdeplot(x)
plt.title("붓꽃data중 꽃잎 길이 커널 밀도 플롯")
plt.show()
```

```python
sns.displot(x,kde=True,rug=True)
plt.show()
```

```python
sns.pairplot(붓꽃data,hue='species')
plt.title("종으로 시각화")
plt.show()
```

```python
타이타닉_sub_data=타이타닉data.pivot_table(index='class',columns='sex',aggfunc='size') #pivot
타이타닉_sub_data
```

```python
sns.heatmap(타이타닉_sub_data,cmap=sns.light_palette('gray',as_cmap=True),annot=True,fmt='d') #heatmap
#annot=True:수치표시,fmt='d':정수표시
```
