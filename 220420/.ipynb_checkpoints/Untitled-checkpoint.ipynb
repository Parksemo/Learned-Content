```python
import numpy as np
import pandas as pd
data = {
    '이름':["길동",'둘리',np.nan,'또치','희동',np.nan],   #np.nan 비어있는데이터 결측치
    '나이':[40,np.nan,15,np.nan,5,np.nan],
    '성별':['남자',np.nan,'여자','여자','남자',np.nan],
    '시험점수':[np.nan,20,80,10,2,np.nan]
}
df=pd.DataFrame(data,columns=['이름','나이','성별','시험점수'])
df
```

```python
df.isnull()
```

```python
df.isnull().sum()/len(df)
```

```python
df2=df.dropna() #dropna 모든 결측치 제거
df2
```

```python
df #드롭해도 원본유지
```

```python
df3=df.dropna(how='all') #how='all' 행의 모든값이 nan이면 해당열 제거
df3
```

```python
df['수정요망']=np.nan
df
```

```python
df.dropna(axis=1,how='all') #(axis=1,how='all') axis을 이용해 열 축 제거 axis= 0:행,1:열
```

```python
df.dropna(axis=0,thresh=1) #thresh=1 nan이 아닌 값이 1개 이상인 행만 표시
df.dropna(axis=1,thresh=1) #thresh=1 nan이 아닌 값이 1개 이상인 열만 표시
```

```python
df.dropna(thresh=3)
```

```python
df4=df.dropna(thresh=2).dropna(axis=1,thresh=1)
df4
```

```python
df4.fillna(0) #fillna 해당값으로 nan을 채운다
```

```python
df4['이름'].fillna('희선',inplace=True)
df4
```

```python
df4['나이'].fillna(df4['나이'].mean(),inplace=True) #mean() 평균값
df4
```

```python
df4.groupby('성별')['시험점수'].transform('mean')
```

```python
df4['시험점수'].fillna(df4.groupby('성별')['시험점수'].transform('mean'),inplace=True)
df4
```

```python
df
```

```python
ck=df['나이'].fillna(df4['나이'].mean()) #전체 채우기
ck
```

```python
ck=df['시험점수'].fillna(df4.groupby('성별')['시험점수'].transform('mean')) #특정화 채우기
ck
```

```python
end_df=df
end_df.dropna(axis=0,thresh=1,inplace=True)
end_df.dropna(axis=1,thresh=1,inplace=True)
end_df['시험점수'].fillna(end_df.groupby('성별')['시험점수'].transform('mean'),inplace=True)
end_df['나이'].fillna(end_df.groupby('성별')['나이'].transform('mean'),inplace=True)
end_df['이름'].fillna('희선',inplace=True)
end_df.dropna(inplace=True)
end_df
```
