## 전체여행지 중 꿀 여행지

```python
random.seed(1)
장소 = []
평점 = []
리뷰 = []
for i in range(1000):
    장소.append(f'장소{i+1}')
    평점.append(random.random()*5)
    리뷰.append(random.randint(10,3000))
장소_평점_리뷰=pd.DataFrame(list(zip(장소, 평점, 리뷰)) ,columns=['장소','평점','리뷰'])
장소_평점_리뷰.head(10)
```

```python
reivew = int(len(장소_평점_리뷰['리뷰'])*0.1)
reivew
```

```python
reivew_n = sorted(장소_평점_리뷰['리뷰'])[reivew]
reivew_n
```

```python
reivew_low = 장소_평점_리뷰[장소_평점_리뷰['리뷰'] <= reivew_n]
reivew_low
```

```python
score_n = sorted(reivew_low['평점'],reverse=True)[9]
score_n
```

```python
result = reivew_low[reivew_low['평점'] >= score_n]
result
```

## 키워드 여행지 중 꿀 여행지

```python
random.seed(10)
장소 = []
평점 = []
리뷰 = []
for i in range(10):
    장소.append(f'장소{i+1}')
    평점.append(random.random()*5)
    리뷰.append(random.randint(10,3000))
키워드_장소_평점_리뷰=pd.DataFrame(list(zip(장소, 평점, 리뷰)) ,columns=['장소','평점','리뷰'])
키워드_장소_평점_리뷰
```

```python
review_max = max(키워드_장소_평점_리뷰['리뷰'])
review_max
```

```python
review_score = []
for i in 키워드_장소_평점_리뷰['리뷰']:
    review_score.append(100-(i/review_max*100))
review_score
```

```python
score_score = []
for i in 키워드_장소_평점_리뷰['평점']:
    score_score.append(i*20)
score_score
```

```python
키워드_장소_평점_리뷰['review_score'] = review_score
키워드_장소_평점_리뷰['score_score'] = score_score
키워드_장소_평점_리뷰
```

```python
result = []
for i in range(len(키워드_장소_평점_리뷰)):
    result.append(review_score[i]/2+score_score[i]/2)
result
```

```python
키워드_장소_평점_리뷰['result'] = result
키워드_장소_평점_리뷰
```

```python
result_max = max(키워드_장소_평점_리뷰['result'])
result_max
```

```python
키워드_장소_평점_리뷰[키워드_장소_평점_리뷰['result'] == result_max].iloc[:,:3]
```
