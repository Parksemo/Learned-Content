```python
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9A%B0%EB%A5%B4%EB%9D%BC%EC%9D%B4%EB%82%98+%EB%9F%AC%EC%8B%9C%EC%95%84')
resp
```

```python
soup = BeautifulSoup(resp.text)
```

```python
news_tags = soup.find_all('div', class_='news_area')
news_tags
```

```python
result_set = []
for tag in news_tags:
    url = tag.get('data-url')
    title = tag.text
    result_set.append({
        'title': title,
        'url': url,
        'desc': title
    })
print(result_set)
```
