```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://search.naver.com/search.naver'
base_params = {
    'where': 'news',
    'query': "우크라이나 러시아",
    'start': 1
}
resp = requests.get(base_url, base_params)

soup = BeautifulSoup(resp.text)
new_list_tags = soup.select('.list_news > .bx')
result_list = []
for news_list_tag in new_list_tags:
    title_tag = news_list_tag.select_one('.news_tit')
    title = title_tag.text
    url = title_tag.get('href')
    desc = news_list_tag.select_one('.dsc_txt_wrap').text
    result_list.append({
        'title': title,
        'url': url,
        'desc': desc
    })
result_list
```

```python
base_url = 'https://search.naver.com/search.naver'
base_params = {
    'where': 'news',
    'query': "우크라이나 러시아",
    'start': 1
}
result_list = []
for i in range(10):
    base_params['start'] = i*10+1
    resp = requests.get(base_url, base_params)

    soup = BeautifulSoup(resp.text)
    new_list_tags = soup.select('.list_news > .bx')
    
    for news_list_tag in new_list_tags:
        title_tag = news_list_tag.select_one('.news_tit')
        title = title_tag.text
        url = title_tag.get('href')
        desc = news_list_tag.select_one('.dsc_txt_wrap').text
        result_list.append({
            'title': title,
            'url': url,
            'desc': desc
        })
```
