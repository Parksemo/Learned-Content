# XML Parsing

```python
import requests
# BeautifulSoup
from bs4 import BeautifulSoup
```

```python
resp = requests.get('https://naver.com')
resp
```

```python
soup = BeautifulSoup(resp.text)
```

```python
group_nav_tag = soup.find('div', class_='group_nav')
group_nav_tag
```

```python
type(group_nav_tag)
```

```python
group_nav_tags = soup.find_all('div', class_='group_nav')
type(group_nav_tags)
```

```python
a_tags = group_nav_tag.find_all('a')
a_tags
```

```python
sample = a_tags[0]
sample
```

```python
# 문자열 가져오기
sample.text
```

```python
sample.get('class')
```

```python
sample.get('data-clk')
```

```python
sample.get('href')
```

```python
result = [{'url': tag.get('href'), 'name': tag.text} for tag in a_tags]
print(result)
```

```python
result_set = []
for tag in a_tags:
    url = tag.get('href')
    name = tag.text
    result_set.append({
        'url': url,
        'name': name
    })
print(result_set)
```
