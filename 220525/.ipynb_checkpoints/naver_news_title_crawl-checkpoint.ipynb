```python
# 라이브러리 import
import pandas as pd
import numpy as np

import chromedriver_autoinstaller
from selenium import webdriver  # 라이브러리(모듈) 가져오라
from selenium.webdriver import ActionChains as AC
from tqdm import tqdm
from tqdm.notebook import tqdm
import re
from time import sleep
import time

# 검색어
search = "손흥민"

# 뉴스 제목에 포함될 필수 단어
essential = "득점왕"

# 워닝 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
options = webdriver.ChromeOptions()
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path, options=options)
driver.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}".format(search))
time.sleep(3)
```

```python
# 옵션 클릭
option_bt = driver.find_element_by_css_selector(".btn_option._search_option_open_btn")
option_bt.click()
time.sleep(1)
```

```python
# 기간 선택
driver.find_element_by_link_text('1주').click()  # 1시간, 1일, 1주, 1개월, 3개월, 6개월, 1년
```

```python
# 모든 page에서 뉴스 title list 수집하기
page_count = driver.find_elements_by_css_selector('.sc_page_inner .btn')
page_num = len(page_count)
print('페이지 수: ', page_num)
```

```python
# 뉴스 전체 페이지를 돌면서 제목에 essential이 포함된 기사의 title, url 수집하기

url_list = []
title_list = []

for i in range(page_num):
    
    # 해당 page에서 뉴스 title list 수집하기
    things = driver.find_elements_by_css_selector('.news_tit')
    len(things)
    
    # title에 essential이 들어있는 기사 추려내기
    in_title_list = []
    for thing in things:
        title = thing.get_attribute('title')
        if essential in title:    # essential : 필수 단어
            in_title_list.append(thing)
    print('essential이 제목에 포함된 기사 수 :', len(in_title_list))
    if len(in_title_list) == 0:
        break

    time.sleep(2)

    for thing in in_title_list:
        title = thing.get_attribute('href')
        url_list.append(title)
    time.sleep(1)
                
    for thing in in_title_list:
        title = thing.get_attribute('title')        
        title_list.append(title)
        print(title)
    time.sleep(1)     
    print('')

    # page 이동 버튼 누르기
    try:
        driver.find_element_by_css_selector('.btn_next').click()
    except:
        break
        
print('크롤링 완료')
```

```python
df = pd.DataFrame({"title":title_list, "url":url_list})
df
```

```python
# 파일로 저장
df.to_excel("naver_news_keyword_in_title_({}).xlsx".format(search))
```
